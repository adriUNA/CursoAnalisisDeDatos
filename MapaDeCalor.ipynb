```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('/content/drive/MyDrive/steam_games.csv')

#Función que cuenta los juegos gratis y de paga
def contador_juegos(df):
    gratis = df[df['original_price'] == 'Free']
    pago = df[df['original_price'] != 'Free']

    cantidad_gratis = len(gratis)
    cantidad_pago = len(pago)

    return cantidad_gratis, cantidad_pago

# Llama a la función y almacena los resultados en variables respectivas.
cantidad_gratis, cantidad_pago = contador_juegos(df)

# Print los resultados.
print("Cantidad de juegos gratis:", cantidad_gratis)
print("Cantidad de juegos de pago:",cantidad_pago)

#Grafico de barras
sns.barplot(x=['Juegos Gratis', 'Juegos de Pago'], y=[cantidad_gratis, cantidad_pago])
plt.title("Cantidad de Juegos Gratis vs Juegos de Pago")
plt.ylabel("Cantidad")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def ordenar_por_fecha(df):
    # Convierte 'release_date' al formato de fecha de Año/Mes/Día
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

    # Ordena el df por 'release_date'
    ordenado_fecha = df.sort_values(by='release_date')

    # Devuelve el df ordenado con las columnas 'release_date' y 'name'
    return ordenado_fecha[['release_date', 'name']]

# Llama a la función y asigna el resultado
resultado_ordenado = ordenar_por_fecha(df)

# Create a density plot for 'release_date'
sns.kdeplot(data=resultado_ordenado['release_date'], fill=True)

# Customize the plot
plt.title("Density Plot of Release Dates")
plt.xlabel("Release Date")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('/content/School-Age-Digital-Connectivity - Primary.csv')
contingency_table=pd.crosstab(df['Income Group'],df['Region'])

#crear el mapa de calor
plt.figure(figsize=(10,7))
sns.heatmap(contingency_table,annot=True,cmap="coolwarm")

#mostrar el grafico
plt.title('Mapa de Calor de Income Group vs residence-urban')
plt.show
```
